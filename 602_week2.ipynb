Question 3. Analyze the make column.

Comment on what you observe.

Clean the column to reduce the number of unique values to as few as reasonably possible (no hard coding).

Hint: What do you notice if you run this:

```python
makes = df.make.unique().tolist()
makes = [str(x).lower() for x in makes]
sorted(makes)
```

- by using unique function, we removed duplicate values in make column.
- Because of using lower function, Upper case names also changed into lower case. So that same names with different cases returned multiple times.
- We sorted the list, which displays in alphabetical order.

```python
makes = [*set(makes)]
sorted(makes)
```

```python
df['mmr'] = pd.to_numeric(df['mmr'], errors='coerce')
```

```python
from statistics import mode
```

```python
mode = mode(df['mmr'].dropna())
df['mmr'].fillna(mode, inplace=True)
```

```python
df['mmr'].astype('int')
```

```python
correlation = df['mmr'].corr(df['sellingprice'])
```

```python
print(correlation)
```

```python
df.sellingprice = df.sellingprice*2
df
```

```python
correlation1 = df['mmr'].corr(df['sellingprice'])
print(correlation1)
```
