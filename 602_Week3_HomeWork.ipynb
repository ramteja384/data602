```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/msaricaumbc/DS_data/master/ds602/iris.csv')
df.info()
df
```

```python
df.species.value_counts()
```

```python
def predict_one(sepal_length, sepal_width, petal_length, petal_width):
    if petal_length < 2 and petal_width < 1.0:
        return 'Iris-setosa'
    elif petal_length >= 2 and petal_length <= 4.5 and petal_width >= 1.0 and petal_width <= 2:
        return 'Iris-versicolor'
    elif petal_length > 4.5 and petal_width > 2:
        return 'Iris-virginica'
    else:
        return 'Unknown' 
    return 'Iris-setosa'
```

```python
predictions = pd.Series([ predict_one(*line) for line in df.iloc[:, :4].to_numpy()])
accuracy = sum(df.species == predictions) / len(df)
print(f"Accuracy: {accuracy}")
```
