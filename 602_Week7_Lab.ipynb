```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('https://raw.githubusercontent.com/msaricaumbc/DS_data/master/ds602/log_reg/employee-turnover-balanced.csv')
data=pd.DataFrame(df)
data
```

```python
data.isnull().sum()
```

```python
data=pd.get_dummies(data, columns=["frequency_of_travel","department","gender","position","married_or_single"])
y = data['left_company']
X = data.iloc[:, 1:]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=3)
```

```python
y = DecisionTreeClassifier(random_state=3)
y.fit(X_train, y_train)
y_pred = y.predict(X_test)
y_acc = accuracy_score(y_test, y_pred)
y_acc
```

```python
y1 = RandomForestClassifier(random_state=3)
y1.fit(X_train, y_train)
y1_pred = y1.predict(X_test)
y1_acc = accuracy_score(y_test, y1_pred)
y1_acc
```
